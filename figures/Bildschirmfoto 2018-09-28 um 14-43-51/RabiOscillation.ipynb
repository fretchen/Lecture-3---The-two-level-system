```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import pi

%matplotlib inline
```

```python
tlin = np.linspace(0,4,100);
omega = 2*pi;
delta = 1*omega;

Pp = 1/(1+(delta/omega)**2)*np.sin(np.sqrt(delta**2+omega**2)*tlin/2)**2;

xz =2*pi/np.sqrt(delta**2+omega**2)
f, ax = plt.subplots()
ax.plot(tlin, Pp, 'r', lw=2)
ax.axhline(1/(1+(delta/omega)**2), c='k', ls='--')
ax.axvline(xz, c='k', ls='--')
ax.text(1, 0.6, '$\sin^2(\\theta)$', fontsize = 12, color = 'k');
ax.text(0.8, 0.9, '$\\frac{\hbar}{E_+-E_-}$', fontsize = 12, color = 'k');
ax.set_xlabel('t')
ax.set_xlabel('t')
ax.set_xticks([])
ax.set_yticks([0,1])
ax.set_ylim(0,1)
```
