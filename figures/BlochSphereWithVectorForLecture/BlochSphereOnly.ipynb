and with the vector

```python
theta = 0.3*pi;
phi = 0.4*pi;
# Bloch sphere
b = Bloch()
#b.xlabel = ['$\\frac{\\left|0\\right>+\\left|1\\right>}{\\sqrt{2}}$', '']
#b.ylabel = ['$\\frac{\\left|0\\right>+i\\left|1\\right>}{\\sqrt{2}}$', '']
b.xlabel = ['', '']
b.ylabel = ['', '']

# state
x = np.sin(theta)*np.cos(phi);
y = np.sin(theta)*np.sin(phi);
z = np.cos(theta)
vec_init = [x, y, z]
b.add_vectors(vec_init)

# projected
# and for the projection on the equator
Np = 5;
zp = [np.cos(th) for th in np.linspace(0, theta, Np)]
yp = [np.sin(th)*np.sin(phi) for th in np.linspace(0, theta, Np)]
xp = [np.sin(th)*np.cos(phi) for th in np.linspace(0, theta, Np)]
pnts = [xp, yp, zp]
b.add_points(pnts)
b.point_color = ['g']

# and for the rotation around on the equator
Np = 5;
zp = np.zeros(Np)
yp = [np.sin(th) for th in np.linspace(0, phi, Np)]
xp = [np.cos(th) for th in np.linspace(0, phi, Np)]
pnts = [xp, yp, zp]
b.add_points(pnts)
b.point_color = ['g']


#b.zlabel = ['$\\left|e\\right>$', '$\\left|g\\right>$']
b.show()
b.save(name = 'BlochSphereWithVector.pdf', format = 'pdf' )
```
