```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import pi

%matplotlib inline
```

```python
deltas = np.linspace(-5,5,100);

omega = 1;

Ep = 1/2*np.sqrt(deltas**2+omega**2);
Em = -1/2*np.sqrt(deltas**2+omega**2);
```

```python
f, ax = plt.subplots()
ax.plot(deltas, Ep, 'r')
ax.plot(deltas, Em, 'r')
ax.plot(deltas, deltas/2, 'k--')
ax.plot(deltas, -deltas/2, 'k--')
ax.axhline(omega/2)
ax.axhline(-omega/2)
f.savefig('AvoidedCrossing.pdf', bbox = 'tight')
```
